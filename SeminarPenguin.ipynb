*Создать новый столбец height_group в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина. high - высокий(от 42), middle - средний(от 35 до 42), low - низкий(до 35).*

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.loc[penguins['bill_length_mm'].ge(42), 'bill_type'] = 'high'
penguins.loc[penguins['bill_length_mm'].between(35, 42, 'left'), 'bill_type'] = 'middle'
penguins.loc[penguins['bill_length_mm'].lt(35), 'bill_type'] = 'low'
penguins
```

*В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?*

```python
import pandas as pd
import random

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

data.loc[data['whoAmI'] == 'robot','robot'] = 1
data.loc[data['whoAmI'] != 'robot','robot'] = 0
data.loc[data['whoAmI'] == 'human','human'] = 1
data.loc[data['whoAmI'] != 'human','human'] = 0
data[['human','robot']]
     
```
